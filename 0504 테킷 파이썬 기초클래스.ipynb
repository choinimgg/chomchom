```python
# 반복문 while ~ 하는 동안에 : 조건이 True 면 계속 실행

if 10 < 90:  # if 참이면 한 번 실행
  print("True") 
```

```python
while 10 < 90:  ## 참이면 무한반복됨(=무한루프). 계속 돌아감
  print("True")
```

```python
i = 0

while i < 10:
  i += 1

  if i % 2 == 0:
    continue    # 2의 배수는 출력되지 않고 다시 돌아가게 됨 = 무시 / continue 도 if문과 함께 작동
    print(i)

print("반복종료")
```

```python
# 사용자에게 로또를 몇 개 살 건지 숫자 입력
# 1부터 45까지 숫자 중 6개를 랜덤으로 뽑기
# 사용자에게 입력받은 개수만큼 2를 뽑기
# 오름차순으로 정렬해서 출력하기


lotto1 = int(input("로또 몇 장 드릴까요? "))
lotto2 = random.sample(range(1,46),6)
i = 0

while True:
  i += 1
  lotto2.sort()
  print(lotto2)
  if i == lotto1 :
   break

print("로또 종료")
```
