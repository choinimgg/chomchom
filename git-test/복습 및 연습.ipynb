```python
# 함수 안에 있는 변수나 함수는 함수가 끝나면 휘발됩니다. 
# 메모리에서 사라져요.
def f2(x, y):
    def f1(x):
        print('world1')
        return x + x
    print('hello1')
    z = f1(10)
    return x + y + z

# a = f1(1) # 함수 안에 있는 함수는 밖에서 호출할 수 있지 않습니다.
b = f2(3, 4)
print(b)

#print(z) # 함수가 끝났기 때문에 x, y, z에 접근할 수 없습니다.
```

```python
print(input("재밌는 사칙연산 계산을 해볼까요"))

def plus(num1, num2):
	return num1 + num2

def minus(num1, num2):
	return num1 - num2

def multiply(num1, num2):
	return num1 * num2

def divide(num1, num2):
	return num1 / num2

print('plus : ',plus(2,1))
print('minus : ',minus(3,4))
print('multiply : ',multiply(4,5))
print('divide : ',divide(15,5))
```
